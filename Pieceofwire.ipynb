Consider the following parabola $f(x)$, which is drawn below. We can see that this parabola satisfies the following equations:

\begin{align*}
f(0) = 6\\
f(2) = 0\\
f(6) =0
\end{align*}


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.linspace(-2.0, 10.0)
s = 1/2*(t-2)*(t-6)

# Note that using plt.subplots below is equivalent to using
# fig = plt.figure() and then ax = fig.add_subplot(111)
fig, ax = plt.subplots()
ax.plot(t, s)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['left'].set_smart_bounds(True)
ax.spines['bottom'].set_smart_bounds(True)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.set(xlabel='x', ylabel='y',
       title='Figure 1: Target function')


fig.savefig("ex1target.png")
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# The functions that we want to plot
t = np.linspace(-6.0, 6.0) #independent variable
a = 0.25*t**2 #parabola 1 which we call a
b = 0.5*t**2 #parabola 2, b
c = 0.75*t**2 #3,c
d = t**2 #4,d
e = 1.5*t**2 #parabola 5, which we call e

#plotting all five functions on one graph
fig, ax = plt.subplots()
ax.plot(t, a)
ax.plot(t,b)
ax.plot(t,c)
ax.plot(t,d)
ax.plot(t,e)

#adding the x and y axis
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['left'].set_smart_bounds(True)
ax.spines['bottom'].set_smart_bounds(True)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

#Labelling our two axes and giving the plot a title
ax.set(xlabel='x', ylabel='y',
       title='Figure 2: Family of parabolas')

fig.savefig("ex1candidates.png")
#putting it all together
plt.show()
```
