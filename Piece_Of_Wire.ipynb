## Example 1: Problem
In the Figure 1 below, I have drawn a “target” parabola––the
graph of a quadratic polynomial $y = f(x)$––with the
following properties:

\begin{align*}
f(0) &= 6\\
f(2) &= 0\\
f(6) &= 0
\end{align*}

In Figure 2, I have also plotted a number of members of the
family $y = a x^2$ for different values of $a > 0$.

It turns out that there is a unique value of $a$ for which the
two graphs are geometrically the same. That is, if the
graphs were made of a stiff wire and the graph $y = ax^2$
could be picked up and translated (to the right and
then down) it would lie exactly on top of the target
graph. The goal for this example is to find this value of $a$.

### Finding the vertex and plotting to test our predition
I think they will certainly need the location of the
vertex of the target parabola and a good way to find
that is to find its equation. By observation of the target function, we see that the roots are $x = 2$ and $x =
6$, and so the function has the form:
$$y = k(x − 2)(x − 6)$$
for some $k > 0$. Now use the fact that $f(0) = 6$:
$$6 = k(0 − 2)(0 − 6) = 12k$$
And this solves to give $k = \frac{1}{2}$. The target graph is therefore:
$$y = \frac{1}{2}(x-2)(x-6)$$
__Goal:__ Build a script to plot the function $y = \frac{1}{2}(x-2)(x-6)$ and check whether it is the same as the target function in Figure 1 above.

```python
#This is the place they would put their python code.
```

```python
#This is one possible code for the goal above
import math
import matplotlib.pyplot as plt
import numpy as np

#Setting up our variables that we want to plot
x = np.linspace(-2.0, 10.0) #independent variable (range can be determined by looking at the target figure)
y= 0.5*(x-2)*(x-6)#The function  we want to plot
plt.plot(x,y) #Setting up our plot
plt.show() #Displaying/outputting our plot
```

We can now find the vertex. The easiest observation
is that the $x$-coordinate of the vertex is half-way between the
roots, so is $x = 4$. Using this we see that the $y$-coordinate is:
$$y(4) =\frac{1}{2} (4-2)(4-6) = \frac{1}{2}(2)(-2) = -2$$

The vertex is at $(4, –2)$.

### Finding a
Now how do we find the value of $a$? There are
perhaps two ways to begin – one is to move the target
parabola so that its vertex is at the origin and the
other is to move the graph $y=ax^2$ so that its vertex
is at $(4, –2)$.

#### Method 1: Move the target parabola

Imagine the vertex of the target parabola is moved to
the origin. Then consider the root at $x = 6$ ––where
will that move to? Well that point is 2 units to the
right of the vertex and 2 units above the vertex, so
after the graph is moved, it will have x-value 2 and y-value 2. 

We can plug these values into the general equation
\begin{align*}
y &= ax^2\\
2 &= a(2)^2\\
2&=4a \\
\implies a&= \frac{1}{2}
\end{align*}
Hence, the family member is: $y = \frac{1}{2}x^2$.

__ Goal: __ modify the code used to generate the target function above to move it up two units and left four units so that the vertex of the target parabola coincides with the origin.


```python
#This is one possible solution
import math
import numpy as np
import matplotlib.pyplot as plt

# The functions that we want to plot (same as above)
x = np.linspace(-2.0, 10.0) #independent variable
y = 0.5*(x-2)*(x-6)

#some additional code required to plot the x and y axes
fig,ax = plt.subplots()
ax.plot(x-4, y+2) # THIS IS THE CHANGE WE NEED TO MAKE TO SHIFT OUR GRAPH

#setting up our axes (same as above)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['left'].set_smart_bounds(True)
ax.spines['bottom'].set_smart_bounds(True)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

#Giving our graph a title and labelling axes
ax.set(xlabel='x', ylabel='y',
       title='Target function after shift to origin')#NEW TITLE

#displaying the plot (same as above)
plt.show()
```

### Method 2: Move the graph $y = ax^2$
Take the unknown graph $y = ax^2$ and move it 4
units right and 2 units down.

\begin{align*}
\text{Start} & &&y=ax^2\\
\text{Move right 4 units} &  &&y= a(x-4)\\
\text{Move down 2 units} & &&y=a(x-4) - 2
\end{align*}

There are lots of ways to go from here. For example
use the fact that $x = 6$ has to be a root.
Set $x = 6$, $y = 0$: 
\begin{align*}
0 &= a(6 − 4)^2 - 2\\
0 &= 4a − 2
\end{align*}

and that gives us $a = \frac{1}{2}$ as before.

__ Goal: __ Using a similar code as the one above, move the function $y = \frac{1}{2}x^2$ right 4 units and down 2 units.

```python
#This is where the students would enter their code
```

```python
#This is one possible solution
import math
import numpy as np
import matplotlib.pyplot as plt

# The functions that we want to plot 
x = np.linspace(-6.0, 6.0) #independent variable
y = 0.5*x**2 # WE HAVE CHANGED THIS FUNCTION

#some additional code required to plot the x and y axes
fig,ax = plt.subplots()
ax.plot(x+4, y-2) #SHIFTING OUR GRAPH right 4, down 2

#setting up our axes (same as above)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['left'].set_smart_bounds(True)
ax.spines['bottom'].set_smart_bounds(True)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

#Giving our graph a title and labelling axes
ax.set(xlabel='x', ylabel='y',
       title='y=1/2 x^2 function after shift to (4,-2)')#NEW TITLE

#displaying the plot (same as above)
plt.show()
```

Now, we are trying to think about how to translate this graph of $y = 1/x$ so that it coincides with our target graph in the figure.
Thinking graphically, here’s what seems ought to work.
1. Flip it vertically. For example, reflect it in the $x$ axis.
2. Lift it so that the asymptote is $y = 4$.
3. Translate it so that it passes through the origin.

Imagine each of these in your mind.  Now we want to do this algebraically. Let’s go through the steps.
### Step 1. Flip it vertically through the x-axis.
Algebraically this changes the sign of y.
So
y = 1/x becomes y = –1/x

__ Goal: __ Plot the function $y = - 1/x$ for $0.25 \leq x \leq 4$. What is the horizontal asymptote of the function?

```python
#This is where the students write their solution
```

```python
#This is one possible code for the goal above
import math
import matplotlib.pyplot as plt
import numpy as np

#Setting up our variables that we want to plot
x = np.linspace(0.25, 4.0) #independent variable (range is given)
y= -1/x #The function  we want to plot
plt.plot(x,y) #Setting up our plot
plt.show() #Displaying/outputting our plot
```

We see that the function plotted above has a horizontal asymptote at $y=0$.


### Step 2: Lift it so that the asymptote is $y = 4$.

Algebraically this adds 4 to $y$. So
$y = –1/x$ becomes $y = 4 – 1/x$ 

__Goal:__ Plot the function $y = 4 - 1/x$ for $0.25 \leq x \leq 4$. Check that the horizontal asymptote of this function is at $y=4$.

```python
#This is where the students write their solution
```

```python
#This is one possible code for the goal above
import math
import matplotlib.pyplot as plt
import numpy as np

#Setting up our variables that we want to plot
x = np.linspace(0.25, 4.0) #independent variable (range is given)
y= 4 -1/x #The function  we want to plot
plt.plot(x,y) #Setting up our plot
plt.show() #Displaying/outputting our plot
```

### Step 3: 3. Translate it so that it passes through the origin.

We want to translate the graph that we plotted in step 2 above by a small amount to the left, so
that it passes through the origin. But how much do we move it? 

We have to find out where it crosses the $x$-axis. That’s where $y = 0$:
\begin{align*}
4 – 1/x &= 0\\
4 &= 1/x\\
x &= 1/4
\end{align*}
Thus we have to translate it 1/4 to the left. How do
we do that?––by adding 1/4 to $x$:
\begin{align*}
y &= 4 − 1/x\\
y &= 4 − \frac{1}{x + 1/4}
\end{align*}
Now simplify. Multiply top and bottom of the
fraction by 4:
\begin{align*}
y &= 4 - \frac{4}{4x + 1}
\end{align*}
Use a common denominator:
\begin{align*}
y&= \dfrac{4(4x+1) - 4}{4x+1}\\
y&=\dfrac{16x}{4x+1}
\end{align*}


__Goal:__ Plot the function $y = \frac{16x}{4x-1}$. Check to see if it matches the target graph.

```python
#This is where the students write their solution
```

```python
#This is one possible code for the goal above
import math
import matplotlib.pyplot as plt
import numpy as np

#Setting up our variables that we want to plot
x = np.linspace(0, 4.0) #independent variable (RANGE NEEDS TO BE CHANGED TO INCLUDE x=0)
y= (16*x)/(4*x + 1) #The function  we want to plot
plt.plot(x,y) #Setting up our plot
plt.show() #Displaying/outputting our plot
```
